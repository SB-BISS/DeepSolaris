```python
import cv2
from os import listdir
import numpy as np

path_negatives="./tim_samples/Negatives/"

path_positives="./tim_samples/Positives/"

img_rows =224
img_cols = 224


#first 100
imagesListNegatives = listdir(path_negatives)

#second 100
imagesListPositives = listdir(path_positives)


labels = np.zeros((len(imagesListNegatives)+len(imagesListPositives)))
labels[len(imagesListPositives):len(imagesListNegatives)+len(imagesListPositives)] = 1


ListImgs = []
for i in imagesListNegatives:
    img = cv2.imread(path_negatives+str(i))
    img = cv2.resize(img,(224,224),interpolation = cv2.INTER_CUBIC)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ListImgs.append(img)

for i in imagesListPositives:
    img = cv2.imread(path_positives+str(i))
    img = cv2.resize(img,(224,224),interpolation = cv2.INTER_CUBIC)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ListImgs.append(img)

NListIm = np.asarray(ListImgs)
#NListIm =NListIm.reshape(NListIm.shape[0],3,img_rows,img_cols) # preparing for tensor...theano version, use above with tf
#We have to extract features from everything then split in training and testing


```

2) Extract the features using an intermediate layer, then save the intermediate results

```python
from keras.models import Model
from sklearn.decomposition import PCA
from matplotlib import pyplot

layer_name = "flatten10"
#layer_name = "fire8/squeeze1x1"
intermediate_layer_model = Model(inputs=model.input,
                                 outputs=model.get_layer(layer_name).output)
intermediate_layer_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) #need to compile but
#not important with what

feature_extracted_negatives_positives = []
for i in range(0,len(NListIm)):
    print i
    feature_extracted_negatives_positives.append(intermediate_layer_model.predict(np.array([NListIm[i]])).flatten())
```

```python
np.shape(feature_extracted_negatives_positives)
```

3) Print the first 2 components of the PCA

```python
from matplotlib import pyplot as plt

table_negatives_positives = []

for i in feature_extracted_negatives_positives:
    table_negatives_positives.append(i.flatten())

#print(table_cats_dogs)
    
    
pca = PCA(n_components=2)
PCA_Solar = pca.fit_transform(table_negatives_positives)

plt.plot(PCA_Solar[labels==1,0],PCA_Solar[labels==1,1],"*r")
plt.plot(PCA_Solar[labels==0,0],PCA_Solar[labels==0,1],"*b")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

solar_train, solar_test, labels_train, labels_test = train_test_split(feature_extracted_negatives_positives, labels, test_size=0.5,random_state=1234)

solar_train_n = scaler.fit_transform(np.squeeze(solar_train,axis=1))
solar_test_n =scaler.transform(np.squeeze(solar_test,axis=1))


neigh = KNeighborsClassifier(n_neighbors=3)

neigh.fit(solar_train_n, labels_train) 

prediction= neigh.predict(solar_test_n)

print(classification_report(labels_test,prediction))
cm= confusion_matrix(labels_test,prediction)

cm

```

```python
from keras.layers import Dense,Dropout, Flatten
from keras.models import Sequential
import keras

# build a classifier model to put on top of the convolutional model
top_model = Sequential()
top_model.add(Dense(128, input_dim=1000))
top_model.add(Dense(256, activation='relu'))
top_model.add(Dropout(0.5))
top_model.add(Dense(1, activation='sigmoid'))


# compile the model with a SGD/momentum optimizer
# and a very slow learning rate.
top_model.compile(loss='binary_crossentropy',
              optimizer=keras.optimizers.rmsprop(lr=1e-4),
              metrics=['accuracy'])

top_model.fit(np.asarray(solar_train_n),np.asarray(labels_train), epochs=50, batch_size=32)
```

```python
predictions = top_model.predict(solar_test_n)

prediction = [round(pred) for pred in predictions]

print(classification_report(labels_test,prediction))
cm= confusion_matrix(labels_test,prediction)

cm

```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=1234, n_estimators=100)
clf.fit(solar_train_n, labels_train)
prediction= clf.predict(solar_test_n)


print(classification_report(labels_test,prediction))
cm= confusion_matrix(labels_test,prediction)

cm
```

```python
from sklearn import svm

clf = svm.SVC(kernel='linear', degree=3)
clf.fit(solar_train_n, labels_train)
prediction= clf.predict(solar_test_n)


print(classification_report(labels_test,prediction))
cm= confusion_matrix(labels_test,prediction)

cm
```

```python
from sklearn.metrics import roc_curve, auc

y_test=labels_test
y_score=prediction

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

fpr, tpr, _ = roc_curve(y_test, y_score)
roc_auc = auc(fpr, tpr)
print(roc_auc)
# Compute micro-average ROC curve and ROC area
#fpr["micro"], tpr["micro"], _ = roc_curve(y_test.ravel(), y_score.ravel())
#roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```
